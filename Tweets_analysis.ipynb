# **Data Pre-Processing** : Clean Text Data to **TF-ID Vectorization**

## Import and Download the necessary NLTK data
*   **punkt** - Sentence tokenizer for English.
*   **stopwords** - Collection of common words (e.g., "the," "a," "is") often removed in text processing.
*   **wordnet** - Lexical database for English, used for finding synonyms, antonyms, and other semantic relationships between words.
*   **punkt_tab** - Sentence tokenizer for tab-separated text.

```python
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import nltk

# Download necessary NLTK data (if not already installed)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
print("NLTK data downloaded!")
```

## **Lemmatizer** and **Stopwords**

```python
# Initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
print("Lemmatizer and stopwords initialized!")
```

## **Cleaning Techniques**

*   Remove URLs
*   Remove Mentions
*   Remove Special Chars
*   Remove Stopwords
*   LowerCase Text
*   Tokenise Text
*   Lemmatize

```python
# Function to clean the 'Content' column for transformation
def clean_text(text):
    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove mentions (@username)
    text = re.sub(r'@\w+\b', '', text)
    # Remove mentions (including those preceded by "RT")
    # text = re.sub(r'(?:^RT\s+|^\s*RT\s+|\s+RT\s+)@\w+\b', '', text)

    # Remove special characters and numbers (except spaces)
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Convert to lowercase
    text = text.lower()

    # Tokenize the text
    words = word_tokenize(text)

    # Remove stopwords and lemmatize
    cleaned_words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]

    # Join the cleaned words back into a single string
    cleaned_text = ' '.join(cleaned_words)

    return cleaned_text
```

### **Apply Above Changes**

*   Creates a new column named `'Cleaned Content'` containing the cleaned text.

```python
# Apply the cleaning function to the 'Content' column
df['Cleaned Content'] = df['Content'].apply(clean_text)
```

**Show the changes made**

```python
# Print the first few rows of the DataFrame with the cleaned text
print(df[['Content', 'Cleaned Content']].head())
```

## **TF-ID Vectorization**

*   Initializes the `TfidfVectorizer` object with `max_features` set to `5000` to limit the number of features.
*   Transforms the `'Cleaned Content'` column of your DataFrame `df` into numerical vectors using the fitted `TfidfVectorizer` object.

```python
# Initialize TfidfVectorizer to transform text data
vectorizer = TfidfVectorizer(max_features=5000)
print("TfidfVectorizer initialized!")

# Transform the 'Cleaned Content' column into numerical vectors
X_transformed = vectorizer.fit_transform(df['Cleaned Content'])
print("Cleaned Content transformed into numerical vectors!")
```

### Tranformed Feature Matrix

*   The number of rows `(documents)` and number of features `(terms)`.

```python
# Print the transformed feature matrix (shape of the matrix and some sample feature names)
print("Transformed feature matrix:")
print(X_transformed.shape)
```

### **Feature names & Transformed Dense Matrix**

*   **Feature Names:** You see the list of unique words or terms that the model identified as important features in the text data.
*   **Dense Matrix:** You see the numerical representation of the documents, where each row corresponds to a document and each column corresponds to a feature, and the values represent the TF-IDF scores.

```python
# Optional: print the feature names (terms in the vocabulary)
print("\nFeature names (vocabulary):")
print(vectorizer.get_feature_names_out())

# You can access the transformed matrix as a dense array (if you want to visualize it)
X_dense = X_transformed.toarray()
print("\nTransformed dense matrix (first 5 rows):")
print(X_dense[:5])
```
